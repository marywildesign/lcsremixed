```python
# Open the file and read all lines into a list
with open("yourfile.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()

# lines is now a list where each element is one line from the file
for line in lines:
    print(line.strip())  # strip() removes the newline at the end
```
